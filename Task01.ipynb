```python
import pandas as pd
df = pd.read_csv('mel.csv')
df
```

```python
newdf = df.drop(columns = ['Address','Date','Postcode','YearBuilt','Lattitude','Longtitude'])
newdf.dropna(inplace=True)
newdf
```

```python
newdf = pd.get_dummies(newdf,dtype=int)
newdf
```

```python
x = newdf.drop(columns = ['Price'])
y = newdf['Price']
x
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33)
```

```python
mod = LinearRegression()
mod.fit(x_train,y_train)
```

```python
mod.score(x_train,y_train)
```

```python
mod.score(x_test,y_test)
```

```python
from sklearn.linear_model import Lasso
mod1 = Lasso(alpha = 50,max_iter=100,tol=0.1)
mod1.fit(x_train,y_train)
```

```python
mod1.score(x_train,y_train)
```

```python
mod1.score(x_test,y_test)
```

```python
from sklearn.linear_model import Ridge
mod2 = Ridge(alpha = 50,max_iter=100,tol=0.1)
mod2.fit(x_train,y_train)
```

```python
mod2.score(x_train,y_train)
```

```python
mod2.score(x_test,y_test)
```
